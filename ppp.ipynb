```python
from copy import deepcopy
import  numpy as np
from matplotlib import pyplot as plt
np.random.seed(1)
r = np.zeros(shape=(101,101))
for t in range(2):
    a = np.array([[[1,2],[2,1]],[[2,1],[1,2]]])
    u = np.zeros(shape=(101,101))
    v = np.zeros(shape=(101,101))
    k = np.random.randint(0, 2, size=(101,101))
    u[50,50] = 10
    for n in range(100):
        for i in range(1,100):
            for j in range(1,100):
                v[i,j] = (a[k[i,j]][0, 0] * (u[i+1, j] + u[i-1,j]) +
                         (a[k[i,j]][0, 1] + a[k[i,j]][1, 0]) / 4 * (u[i+1, j+1] + u[i-1,j-1] -u[i+1,j-1]-u[i-1,j+1]) +
                          a[k[i,j]][1, 1] * (u[i,j+1] + u[i,j-1])) / (2*(a[k[i,j]][0,0]+a[k[i,j]][1,1]))
        u = deepcopy(v)
        u[50,50] = 10
    r = r + u


x_values = np.linspace(0,10,100)
y_values = np.linspace(0,10,100)
x, y = np.meshgrid(x_values, y_values)
plt.imshow(r / 10)
```
