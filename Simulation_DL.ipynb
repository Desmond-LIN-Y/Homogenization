```python
import numpy as np
from copy import deepcopy
from matplotlib import pyplot as plt
```

```python
def simulate(source_size):
    a = np.array([[[2, 1], [1, 2]], [[20, 1], [1, 1]]])
    mesh_size = 16 * source_size
    u = np.zeros(shape=(mesh_size, mesh_size))
    v = np.zeros(shape=(mesh_size, mesh_size))
    
    def source():
        u[(mesh_size // 2 - source_size) : (mesh_size // 2 + source_size), (mesh_size // 2 - source_size) : (mesh_size // 2 + source_size)] = 10 * np.ones(shape=(source_size * 2, source_size * 2))

    source()

    for n in range(1000):
        for i in range(1, mesh_size - 1):
            for j in range(1, mesh_size - 1):
                p=a[(i + j) % 2]
                v[i,j] = (p[0, 0] * (u[i + 1, j] + u[i - 1,j]) +
                         (p[0, 1] + p[1, 0]) / 4 * (u[i + 1, j + 1] + u[i - 1,j - 1] - u[i + 1,j - 1] - u[i - 1,j + 1]) +
                          p[1, 1] * (u[i, j + 1] + u[i, j - 1])) / (2*(p[0, 0]+p[1, 1]))
        u = deepcopy(v)
        source()
    return u

```

```python
u = simulate(1)
#Plots
x_values = np.linspace(0, 10, 100)
y_values = np.linspace(0, 10, 100)
x, y = np.meshgrid(x_values, y_values)

fig = plt.imshow(u)
plt.axis('off')
plt.savefig("periodic_1.png",pad_inches = 0,bbox_inches='tight')
plt.show()
```

```python
def simulate_random(source_size):
    np.random.seed(1)
    a = np.array([[[2, 1], [1, 2]],[[20, 1], [1, 1]]])
    mesh_size = 16 * source_size
    u = np.zeros(shape=(mesh_size, mesh_size))
    v = np.zeros(shape=(mesh_size, mesh_size))
    k = np.random.choice([0,1], size=(mesh_size, mesh_size),p=(0.5,0.5))
    
    def source():
        u[(mesh_size // 2 - source_size) : (mesh_size // 2 + source_size), (mesh_size // 2 - source_size) : (mesh_size // 2 + source_size)] = 5 * np.ones(shape=(source_size * 2, source_size * 2))

    source()
    
    for n in range(1000):
        for i in range(1, mesh_size - 1):
            for j in range(1, mesh_size - 1):
                v[i,j] = (a[k[i,j]][0, 0] * (u[i+1, j] + u[i-1,j]) +
                         (a[k[i,j]][0, 1] + a[k[i,j]][1, 0]) / 4 * (u[i+1, j+1] + u[i-1,j-1] -u[i+1,j-1]-u[i-1,j+1]) +
                          a[k[i,j]][1, 1] * (u[i,j+1] + u[i,j-1])) / (2*(a[k[i,j]][0,0]+a[k[i,j]][1,1]))
        u = deepcopy(v)
        source()

    return u

```

```python
u = simulate_random(1)
#Plots
x_values = np.linspace(0,5,100)
y_values = np.linspace(0,5,100)
x, y = np.meshgrid(x_values, y_values)
plt.axis('off')
plt.imshow(u)
plt.colorbar()
plt.savefig("random_1.png",pad_inches = 0,bbox_inches='tight')
plt.show()
```
