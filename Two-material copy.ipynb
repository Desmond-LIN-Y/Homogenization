```python
# Imports
import numpy as np
from copy import deepcopy
from matplotlib import pyplot as plt
```

```python
# Global Settings

# Dimension of simulation
dim=100

# We assume two material exists, one with resistance 1, another decided by variable res_2
res_2=100
grid=np.zeros((dim,dim))
center=int((dim-1)/2)
grid[center,center]=6000
```

```python
# Functions
def coefs(i,j):
    '''
    # RU arrow
    if (i+j)%2==0:
        #up,left,down,right
        return (1,1,res_2,res_2)
    else:
        return (res_2,res_2,1,1)
        '''
    # LU arrow
    if (i+j)%2==0:
        #up,left,down,right
        return (1,res_2,res_2,1)
    else:
        return (res_2,1,1,res_2)

def update(i,j):
    #gives updated value of i,j point
    up=grid[i-1,j]
    left=grid[i,j-1]
    down=grid[i+1,j]
    right=grid[i,j+1]
    co=coefs(i,j)
    val=(
        up*co[1]*co[2]*co[3]
        +left*co[0]*co[2]*co[3]
        +down*co[0]*co[1]*co[3]
        +right*co[0]*co[1]*co[2]
    )/(
        co[1]*co[2]*co[3]
        +co[0]*co[2]*co[3]
        +co[0]*co[1]*co[3]
        +co[0]*co[1]*co[2]
        )

    return val
```

```python
newgrid=deepcopy(grid)
for time in range(101):
    for i in range(1,dim-1):
        for j in range(1,dim-1):
            newgrid[i,j]=update(i,j)
    newgrid[center,center]=grid[center,center]
    grid=deepcopy(newgrid)



#a=[i for i in range(dim)]
#x,y=np.meshgrid(a,a)
#plt.matshow(grid, cmap=plt.cm.gray)
#plt.show()

x_values = np.linspace(0,10,100)
y_values = np.linspace(0,10,100)
x, y = np.meshgrid(x_values, y_values)
```
